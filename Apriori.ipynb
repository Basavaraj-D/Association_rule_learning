### what is association rule learning?

Association rule learning is a type of unsupervised learning technique that checks for the dependency of one data item on another data item and maps accordingly so that it can be more profitable. It tries to find some interesting relations or associations among the variables of dataset.

In an even simpler terms we can say ***Those who bought that also bought this*** the below image can summarise association rule learning.

About the dataset:
- our dataset contains all the products bought in  a supermarket by various customers

### Importing libraries and reading the dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv',header =None)
dataset.shape
```

- we have around 7501 datapoints with 20 features

```python
df = dataset.copy()
df.head()
```

```python
df.isnull().sum().sum()
```

- we have a lot of missing values present in our dataset which is quite acceptable since not every customer buys everything,but we do not need to deal with missing values since these missing values are what will help us draw an association

```python
transactions =[]
for i in  range(0,7501):
  transactions.append([str(df.values[i,j]) for j in range(0,20)])
```

- we will convert everything into string since aproyri model takes only string

### Training the Apriori model on the dataset

```python
from apyori import apriori
rules = apriori(transactions=transactions,min_support =0.003,
                min_confidence=0.2,min_lift =3,min_length =2,
                max_length=2)
```

- min_support = we want to see a product atleast 3 times a day  in a week over the total transactions
3*7/7501 = 0.003

- min_confidence = confidence with which we can say the product at the right  will contain the product at the left.

- max_and min length = since we want to give offer as buy one get 1 free which could vary based on the business problem

- In general the best min_lift is 3 but we can optimise based on the preference of the business problem

### Displaying the results coming directly from the output of the apriori function

```python
results = list(rules)
results
```

### Displaying the results well organised into a pandas dataframe

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```

```python
resultsinDataFrame
```

```python
resultsinDataFrame.nlargest(n=10,columns='Lift')
```
