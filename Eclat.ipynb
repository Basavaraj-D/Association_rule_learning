Unlike Apriori we use Eclat when we have a smaller dataset and in Eclat we dont have any sort of confidence ,we just have support which is <br>
- support = transactions containing i / transactions<br>

in Eclat unlike apriori we are just concerned in how many sets both products occured.<br>
Example:<br>
out of the total people who watched movie A how many people also watched Movie B.In here we are not concerned with any sort of confidence as such.

### Importing libraries and dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv',header=None)
dataset.shape
```

```python
df = dataset.copy()
df.head()
```

```python
transactions = []
for i in range(0,7501):
  transactions.append([str(df.values[i,j]) for j in range(0,20)])
```

### Training the Eclat model on the dataset

```python
from apyori import apriori
rules = apriori(transactions=transactions,min_support =0.003,
                  min_confidence=0.2,min_lift=3,min_length=2,max_length=2)

# we can remove min_lift and min_confidence however we will keep it since it does not hinder our process in anyway
```

### Displaying the results 

```python
results = list(rules)
results
```

```python
def inspect(results):
  lhs = [tuple(result[2][0][0])[0] for result in results]
  rhs = [tuple(result[2][0][1])[0] for result in results]
  supports = [result[1] for result in results]
  return list(zip(lhs,rhs,supports))
resultsindataframe = pd.DataFrame(inspect(results),
                                  columns=['Product 1','Product 2','Support'])
```

```python
resultsindataframe.nlargest(n=10,columns='Support')
```
